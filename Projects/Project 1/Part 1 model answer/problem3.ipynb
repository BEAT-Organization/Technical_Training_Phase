```python
from skimage import color   #used to convert to grey scale
from skimage import io      #read/write images
from skimage import util    #used to invert grey scale images

import matplotlib.pyplot as plt     #used in plotting images
```

```python
img = io.imread('vegetables.jpg')
plt.imshow(img)
```

```python
grey_img = color.rgb2gray(img)
plt.imshow(grey_img, cmap='gray')
```

```python
inv_img = util.invert(grey_img)
plt.imshow(inv_img, cmap='gray')
```

```python
plt.figure(figsize=(20,10))

plt.subplot(1, 3, 1)
plt.imshow(img)
plt.title('Original Image')

plt.subplot(1, 3, 2)
plt.imshow(grey_img, cmap='gray')
plt.title('Grey Image')

plt.subplot(1, 3, 3)
plt.imshow(inv_img, cmap='gray')
plt.title('Inverted Image')
```

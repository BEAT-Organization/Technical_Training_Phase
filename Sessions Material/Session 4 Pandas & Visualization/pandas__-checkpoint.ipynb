### Skip The First Row And Change The Index To Col 1

```python
df = pd.read_csv('olympics.csv', index_col = 0 , skiprows = 1 )
df.head()
```

### Change Columns Names

```python
for col in df.columns:
    if col[:2]=='01': #col[what I want to check if it exists to change]
        df.rename(columns={col:'Gold' + col[4:]}, inplace=True) #col[what I want to keep and add to it its type]
    if col[:2]=='02':
        df.rename(columns={col:'Silver' + col[4:]}, inplace=True)
    if col[:2]=='03':
        df.rename(columns={col:'Bronze' + col[4:]}, inplace=True)
    if col[:1]=='№':
        df.rename(columns={col:'#' + col[1:]}, inplace=True) 

df.head()
```

### Find The Uniqe Values

```python
x = df['Gold'].unique() #the uniques of column
x.sort()
print(x)

#  that is not the only way => drop_duplicates => the uniques of the df for this column
df["Gold"].drop_duplicates()
df["Gold"].sort_values().tail()
```
